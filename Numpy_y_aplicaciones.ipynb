# Introducción a la librería Random de Python

La librería `random` en Python proporciona funciones para generar números aleatorios, seleccionar elementos de una secuencia al azar, y realizar operaciones aleatorias. Es ampliamente utilizada en simulaciones, pruebas aleatorias, y aplicaciones donde se requieren valores aleatorios.

En este cuaderno, exploraremos las principales funcionalidades de la librería `random` a través de ejemplos prácticos y ejercicios.

## Generación de Números Aleatorios

### Números Enteros
La función `random.randint(a, b)` genera un número entero aleatorio `N` tal que `a <= N <= b`.

### Números de Punto Flotante
La función `random.uniform(a, b)` genera un número de punto flotante aleatorio `N` tal que `a <= N <= b`.

### Ejemplos

```
# Número entero entre 1 y 10
randomInt = random.randint(1, 10)
print(f"Número entero aleatorio entre 1 y 10: {randomInt}")

# Número de punto flotante entre 0 y 1
randomFloat = random.uniform(0, 1)
print(f"Número de punto flotante aleatorio entre 0 y 1: {randomFloat}")
```

## Selección de Elementos Aleatorios

### Elemento Aleatorio de una Lista
La función `random.choice(seq)` devuelve un elemento aleatorio de la secuencia `seq`.

### Muestra Aleatoria de una Lista
La función `random.sample(seq, k)` devuelve una lista de `k` elementos seleccionados aleatoriamente de la secuencia `seq`.

### Ejemplos

```
# Lista de ejemplo
elements = ['a', 'b', 'c', 'd', 'e', 'f', 'g']

# Selección de un elemento aleatorio
random_element = random.choice(elements)
print(f"Elemento aleatorio de la lista: {random_element}")

# Selección de una muestra de 3 elementos aleatorios
random_sample = random.sample(elements, 3)
print(f"Muestra aleatoria de 3 elementos: {random_sample}")
```

## Mezcla de una Lista

La función `random.shuffle(seq)` mezcla los elementos de la lista `seq` en su lugar.

### Ejemplo

```
# Lista de ejemplo
numbers = [1, 2, 3, 4, 5]

# Mezcla de la lista
random.shuffle(numbers)
print(f"Lista mezclada: {numbers}")
```

## Generación de Secuencias Aleatorias

### Números Aleatorios
La función `random.random()` devuelve un número de punto flotante aleatorio `N` tal que `0 <= N < 1`.

### Números Aleatorios con Distribución Normal
La función `random.gauss(mu, sigma)` genera un número aleatorio con una distribución normal con media `mu` y desviación estándar `sigma`.

### Ejemplos

```
# Número de punto flotante aleatorio entre 0 y 1
random_float = random.random()
print(f"Número de punto flotante aleatorio entre 0 y 1: {random_float}")

# Número aleatorio con distribución normal (media=0, desviación estándar=1)
random_gaussian = random.gauss(0, 1)
print(f"Número aleatorio con distribución normal (media=0, desviación estándar=1): {random_gaussian}")

# TODO: Verificar que la distribución se aleatoria
# Histograma
```

### Matrices en Python (Listas de listas)

```
import numpy as np
import time

n = 1000

start_time_list = time.time()
# Generar matrices de 10x10 con listas de listas
matrix_list1 = [[np.random.randint(0, 100) for _ in range(n)] for _ in range(n)]
matrix_list2 = [[np.random.randint(0, 100) for _ in range(n)] for _ in range(n)]

# Sumar matrices con listas de listas y medir el tiempo
sum_matrix_list = [[matrix_list1[i][j] + matrix_list2[i][j] for j in range(n)] for i in range(n)]
end_time_list = time.time()
time_list = end_time_list - start_time_list
print("Tiempo total Listas", time_list, "seg")
```

```

# Generar matrices de nxn con NumPy
start_time_np = time.time()
matrix_np1 = np.random.randint(0, 100, (n, n))
matrix_np2 = np.random.randint(0, 100, (n, n))

# Sumar matrices con NumPy y medir el tiempo
sum_matrix_np = matrix_np1 + matrix_np2
end_time_np = time.time()
time_np = end_time_np - start_time_np
print("Tiempo total Numpy", time_np, "seg")

# Calcular acceleración (speedUP)
print("Aceleración", time_list / time_np)
```

```
# Ejercicio 4 1 en diagonal principal y traza de una matrix
import random as rn


def calcularTraza(A):
    """
    Input: Matriz
    Output: Traza de la matriz de entrada
    """
    traza = 0
    for i in range(len(A)):
        traza += A[i][i]
    return traza

def agregarElementos(A):
    for i in range(len(A)):
        #A[i][i] = 1
        A[i][i] = rn.randint(1, 10)
    return A


zeros = np.zeros((10,10), dtype=np.float32)
print("Matriz Original: \n" + str(zeros))

zeros = agregarElementos(zeros)
print("Matriz Original: \n" + str(zeros))

print("La traza de la matriz es:", calcularTraza(zeros))

```

# Procesamiento de Imágenes con NumPy

NumPy es una herramienta poderosa para el procesamiento de imágenes debido a su capacidad para manejar eficientemente matrices multidimensionales. A continuación se presentan algunos ejemplos que demuestran cómo utilizar NumPy para realizar diversas operaciones de procesamiento de imágenes.

## Ejemplo 1: Cargar y Mostrar una Imagen
Utilizaremos la librería `Pillow` (PIL) para cargar una imagen y convertirla en un array de NumPy para su manipulación.

```
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

# Cargar la imagen
image = Image.open('/content/lena.jpeg')

# Convertir la imagen a un array de NumPy
image_np = np.array(image)

# Mostrar la imagen original
plt.imshow(image_np)
plt.title('Imagen Original')
plt.axis('off')
plt.show()
```

## Ejemplo 2: Convertir a Escala de Grises
Convertir una imagen a escala de grises.

```
# Convertir la imagen a escala de grises
gray_image_np = np.mean(image_np, axis=2).astype(np.uint8)

# Mostrar la imagen en escala de grises
plt.imshow(gray_image_np, cmap='gray')
plt.title('Imagen en Escala de Grises')
plt.axis('off')
plt.show()
```

## Ejemplo 3: Invertir los Colores de la Imagen
Invertir los colores de una imagen.

```
# Invertir los colores de la imagen
inverted_image_np = 255 - image_np

# Mostrar la imagen con colores invertidos
plt.imshow(inverted_image_np)
plt.title('Imagen con Colores Invertidos')
plt.axis('off')
plt.show()
```

## Ejemplo 4: Aplicar un Filtro de Desenfoque
Aplicar un filtro de desenfoque a una imagen utilizando un filtro de convolución.

```
def apply_blur(image_np, kernel_size=5):
    kernel = np.ones((kernel_size, kernel_size)) / (kernel_size ** 2)
    image_blur = np.zeros_like(image_np)

    for i in range(image_np.shape[0] - kernel_size + 1):
        for j in range(image_np.shape[1] - kernel_size + 1):
            for k in range(3):
                image_blur[i + kernel_size // 2, j + kernel_size // 2, k] = np.sum(image_np[i:i+kernel_size, j:j+kernel_size, k] * kernel)

    return image_blur

# Aplicar el filtro de desenfoque
blurred_image_np = apply_blur(image_np)

# Mostrar la imagen desenfocada
plt.imshow(blurred_image_np)
plt.title('Imagen Desenfocada')
plt.axis('off')
plt.show()
```

## Ejemplo 5: Detección de Bordes con el Filtro de Sobel
Aplicar el filtro de Sobel para detectar bordes en una imagen.

```
from scipy.ndimage import sobel

# Convertir la imagen a escala de grises
gray_image_np = np.mean(image_np, axis=2).astype(np.uint8)

# Aplicar el filtro de Sobel
sobel_x = sobel(gray_image_np, axis=0)
sobel_y = sobel(gray_image_np, axis=1)
sobel_magnitude = np.hypot(sobel_x, sobel_y).astype(np.uint8)

# Mostrar la imagen con detección de bordes
plt.imshow(sobel_magnitude, cmap='gray')
plt.title('Detección de Bordes con Sobel')
plt.axis('off')
plt.show()
```

## Ejemplo 6: Redimensionar una Imagen
Redimensionar una imagen utilizando NumPy.

```
def resize_image(image_np, new_width, new_height):
    old_height, old_width = image_np.shape[:2]
    resized_image = np.zeros((new_height, new_width, 3), dtype=np.uint8)

    for i in range(new_height):
        for j in range(new_width):
            x = int(j * old_width / new_width)
            y = int(i * old_height / new_height)
            resized_image[i, j] = image_np[y, x]

    return resized_image

# Redimensionar la imagen a 100x100 píxeles
resized_image_np = resize_image(image_np, 3200, 3200)

# Mostrar la imagen redimensionada
plt.imshow(resized_image_np)
plt.title('Imagen Redimensionada a 1000x1000')
plt.axis('off')
plt.show()
```
